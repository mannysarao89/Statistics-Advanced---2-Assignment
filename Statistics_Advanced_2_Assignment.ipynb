Question 1: What is hypothesis testing in statistics?

Answer: Hypothesis testing is a statistical procedure used to assess whether there is enough evidence in a sample of data to infer that a certain condition holds true for the entire population.

Question 2: What is the null hypothesis, and how does it differ from the alternative hypothesis?

Answer: The null hypothesis is a statement that there is no effect, no difference, or no relationship in the population. It represents the default or status quo assumption.

      Feature	                 Null Hypothesis (H‚ÇÄ)	        Alternative Hypothesis (H‚ÇÅ or Ha)
     --------------------------------------------------------------------------------------------
      Meaning	                 No change or effect	        There is a change or effect

      Symbolic Form (Example)	 Œº = 100 (no difference)	  Œº ‚â† 100 (there is a difference)
      Decision Basis	         Rejected or not rejected	  Never "accepted" directly, only supported

Question 4: What are Type I and Type II errors? Give examples of each.

Answer: Type I and Type II Errors are two kinds of mistakes that can happen in hypothesis testing. They occur when we make the wrong decision based on our sample data.

üîπ Type I Error (False Positive): Rejecting the null hypothesis (H‚ÇÄ) when it is actually true.
- You're saying there's an effect or difference, but in reality, there isn't.
- The probability of making this error is the significance level Œ± (e.g., 0.05 means 5% chance).

> Example: A COVID-19 test wrongly says someone is positive when they are actually not infected.
- H‚ÇÄ: The person is healthy (not infected).
- H‚ÇÅ: The person is infected.
- Type I Error: You reject H‚ÇÄ and say they are infected ‚Äî but they're actually healthy.

üîπ Type II Error (False Negative): Failing to reject the null hypothesis when it is actually false.
- You're saying there‚Äôs no effect or difference, but in reality, there is.
- The probability of this error is Œ≤ (beta).

> Example: A smoke detector fails to go off during a real fire.
- H‚ÇÄ: There is no fire.
- H‚ÇÅ: There is a fire.
- Type II Error: You fail to reject H‚ÇÄ ‚Äî you assume there's no fire, but there is one.

Question 5: What is the difference between a Z-test and a T-test? Explain when to use each.

Answer: Both Z-tests and T-tests are used in hypothesis testing to determine if there is a significant difference between groups or from a known value. However, they are used in different situations based on sample size and whether population standard deviation is known.

üîπ Z-test is used when:
- The sample size is large (n ‚â• 30).
- The population standard deviation (œÉ) is known.
- The data is approximately normally distributed.

üîπ T-test is used when:
- The sample size is small (n < 30).
- The population standard deviation is unknown.
- The data is approximately normally distributed.

Question 6: Write a Python program to generate a binomial distribution with n=10 and p=0.5, then plot its histogram.
(Include your Python code and output in the code box below.)
Hint: Generate random number using random function.

Answer:

```python
import numpy as np
import matplotlib.pyplot as plt

n = 10
p = 0.5
size = 10000

data = np.random.binomial(n=n, p=p, size=size)

plt.hist(data, bins=range(n+2), align='left', edgecolor='black', color='skyblue', rwidth=0.8)
plt.title('Binomial Distribution (n=10, p=0.5)')
plt.xlabel('Number of Successes')
plt.ylabel('Frequency')
plt.xticks(range(n+1))
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

Question 7: Implement hypothesis testing using Z-statistics for a sample dataset in Python. Show the Python code and interpret the results.
sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
50.3, 50.4, 50.0, 49.7, 50.5, 49.9]
(Include your Python code and output in the code box below.)

Answer:

```python
import numpy as np
from scipy.stats import norm

sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
               50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
               50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
               50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

population_mean = 50
alpha = 0.05

sample_mean = np.mean(sample_data)
sample_std = np.std(sample_data, ddof=1)
n = len(sample_data)

z_stat = (sample_mean - population_mean) / (sample_std / np.sqrt(n))

p_value = 2 * (1 - norm.cdf(abs(z_stat)))

print(f"Sample Mean: {sample_mean:.3f}")
print(f"Sample Std Dev: {sample_std:.3f}")
print(f"Z-statistic: {z_stat:.3f}")
print(f"P-value: {p_value:.4f}")

if p_value < alpha:
    print("Reject the null hypothesis: There is a significant difference.")
else:
    print("Fail to reject the null hypothesis: No significant difference.")
```

Question 8: Write a Python script to simulate data from a normal distribution and calculate the 95% confidence interval for its mean. Plot the data using Matplotlib.
(Include your Python code and output in the code box below.)

Answer:

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

np.random.seed(42)
mu = 100
sigma = 15
n = 100

data = np.random.normal(loc=mu, scale=sigma, size=n)

sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)
confidence = 0.95
z_critical = stats.norm.ppf(1 - (1 - confidence) / 2)

margin_of_error = z_critical * (sample_std / np.sqrt(n))
ci_lower = sample_mean - margin_of_error
ci_upper = sample_mean + margin_of_error

print(f"Sample Mean: {sample_mean:.2f}")
print(f"Sample Standard Deviation: {sample_std:.2f}")
print(f"95% Confidence Interval: ({ci_lower:.2f}, {ci_upper:.2f})")

plt.figure(figsize=(10, 6))
plt.hist(data, bins=15, color='skyblue', edgecolor='black', alpha=0.7)
plt.axvline(sample_mean, color='red', linestyle='--', label=f'Mean = {sample_mean:.2f}')
plt.axvline(ci_lower, color='green', linestyle='--', label=f'95% CI Lower = {ci_lower:.2f}')
plt.axvline(ci_upper, color='green', linestyle='--', label=f'95% CI Upper = {ci_upper:.2f}')
plt.title('Histogram of Simulated Normal Data with 95% Confidence Interval')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

Question 9: Write a Python function to calculate the Z-scores from a dataset and
visualize the standardized data using a histogram. Explain what the Z-scores represent in terms of standard deviations from the mean.
(Include your Python code and output in the code box below.)

Answer:

```python
import numpy as np
import matplotlib.pyplot as plt

def calculate_z_scores(data):
    mean = np.mean(data)
    std_dev = np.std(data, ddof=1)
    z_scores = [(x - mean) / std_dev for x in data]
    return z_scores

np.random.seed(42)
data = np.random.normal(loc=100, scale=15, size=100)

z_scores = calculate_z_scores(data)

print("First 5 Z-scores:", np.round(z_scores[:5], 3))
print("Last 5 Z-scores:", np.round(z_scores[-5:], 3))

plt.figure(figsize=(10, 6))
plt.hist(z_scores, bins=15, color='lightcoral', edgecolor='black', alpha=0.7)
plt.title('Histogram of Z-scores (Standardized Data)')
plt.xlabel('Z-score')
plt.ylabel('Frequency')
plt.axvline(0, color='blue', linestyle='--', label='Mean (Z = 0)')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```
